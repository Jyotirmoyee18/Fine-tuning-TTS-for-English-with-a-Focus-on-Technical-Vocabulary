```python
import pdfplumber
import nltk
from nltk.tokenize import sent_tokenize
from transformers import Trainer, TrainingArguments, SpeechT5ForTextToSpeech, SpeechT5Processor
from datasets import Dataset

# Download necessary NLTK resources
nltk.download('punkt')

# Function to extract text from a PDF file
def read_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"  # Extract text from each page
    return text

# Replace 'your_file.pdf' with the path to your PDF file
pdf_text = read_pdf('/content/drive/MyDrive/DATASET FOR TTS/DATASET FOR TTS.pdf')
print(pdf_text)  # Display extracted text

# Split text into sentences
sentences = sent_tokenize(pdf_text)
sentences = [s for s in sentences if s and s.strip()]  # Filter out None or empty sentences

# Load pre-trained model and processor
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")

# Create a Hugging Face Dataset
train_dataset = Dataset.from_dict({"text": sentences})

# Preprocess the dataset
def preprocess_function(examples):
    return processor(text=examples["text"], padding="max_length", truncation=True, return_tensors="pt")

# Apply the preprocessing function to the dataset
train_dataset = train_dataset.map(preprocess_function, batched=True)

# Set up training arguments
training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy="epoch",  # Use 'eval_strategy' instead
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    remove_unused_columns=False,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

# Start fine-tuning
trainer.train()
```
